# Business Context

Apple ï£¿ has just launched its latest flagship phone, the iPhone 14, in September 2022 and is interested in analysing public response to this product. Through sentiment analysis, they would like to receive constructive feedback from public regarding features considered positive and negative which could help them in designing the next generation of iPhones.

## STEP 1: Importing Main Libraries

```python
import nltk #Natural Language Toolkit
nltk.download('punkt')
import re, string, unicodedata
```

## STEP 2: Extracting iPhone 14 Reviews from Several Review Websites

To perform this analysis, articles from four popular websites that reviewed iPhone 14 were extracted into text: 

* Forbes
* Tech Radar
* The Guardian
* Business Insider

These review text files are places in a folder, and from here a corpus, which is a collection of texts, is created.

```python
# Reading iPhone 14 review text files from root directory using Corpus Reader

from nltk.corpus import PlaintextCorpusReader
corpus_root = './Data/iphone_14_reviews'
corpus_reader = PlaintextCorpusReader(corpus_root, '.*', encoding='latin1') 

print(corpus_reader)
```

## STEP 3: Tokenizing
Tokenizing is a process to break text into individual linguistic units.

```python
corpus_tokens = corpus_reader.words() 
print(corpus_tokens[620:635]) #This command is to see the sample of tokens
```

```python
# Examining length of corpus tokens before any pre-process is performed

print (len(corpus_tokens))
```

```python
# Analysing frequency distribution of corpus tokens

from nltk.probability import FreqDist
frequencies = FreqDist(corpus_tokens)
print(frequencies)
```

```python
# Printing 100 most common tokens

print(frequencies.most_common(100))
```

### Observations:
It can be observed from the data that token pre-processing is needed because there are several occurrences of stopwords, punctuations, and digits which does not give significant meaning to the text, so they need to be removed before drawing any conclusions.

## STEP 4: Pre-Processing Corpus Tokens

```python
# Converting the tokens into lowercase

lower_corpus_tokens = [word.lower() 
                       for word in corpus_tokens]
```

```python
# Observing the list of punctuations, digits, and stopwords which could be removed from our corpus tokens 

nltk.download('stopwords')
from nltk.corpus import stopwords

print("Punctuations: " + string.punctuation)
print("Digits: " + string.digits)
print("Stopwords: ")
print(stopwords.words('english'))
```

```python
# Creating a function to handle pre-process and removal of certain list of words

def pre_process_tokens (tokens_list, remove_tokens):
    filtered_tokens = [token
                      for token in tokens_list
                      if not token in remove_tokens]
    return filtered_tokens    
```

```python
# First pre-processing attempt at removing stop words, punctuation, and digits from token list

remove_first = set(stopwords.words('english') + list(string.punctuation) + list(string.digits)) #list of elements to delete
first_filtered_tokens = pre_process_tokens(lower_corpus_tokens,remove_first)

frequencies_filtered = FreqDist(first_filtered_tokens)
print(frequencies_filtered.most_common(100))
```

### Observations:

It can be seen that there is still a requirement to remove the following tokens from the corpus tokens list before analysis can be performed:
* Tokens of the form '\x80' and '\x00' which have some of the most common occurences
* Tokens containing 'Ã¢' which has occurred in different forms multiple times
* Extra punctuations like ').' which has occured 9 times
* Individual letters

```python
# Creating a list of '\x80' and '\x00' tokens to be removed using Regular Expression RE

backslash_x80=[word 
                for word in first_filtered_tokens 
                if re.search('\x80.', word)]
print(backslash_x80)

backslash_x00=[word 
                for word in first_filtered_tokens 
                if re.search('\x00', word)]
print(backslash_x00)
```

```python
# Creating a list of individual letters tokens to be removed

individual_letters = list(string.ascii_lowercase)
print(individual_letters)
```

```python
# Creating a list of meaningless tokens containing 'Ã¢' to be removed using Regular Expressin RE

meaningless_unicode=[word 
                for word in first_filtered_tokens 
                if re.search('Ã¢', word)]
print(meaningless_unicode)
```

```python
# Creating a list of extra punctuations tokens to be removed

extra_punctuations = ['.', '..','...','....','.....','......', ').', '.,','$','Â£']
print(extra_punctuations)
```

```python
# Second pre-processing attempt to remove other meaningless token from our token list

remove_second = set(list(backslash_x80) + list(backslash_x00) + list(individual_letters) + list(meaningless_unicode) + list(extra_punctuations)
)
filtered_tokens = pre_process_tokens(first_filtered_tokens,remove_second)

frequencies_filtered = FreqDist(filtered_tokens)
print(frequencies_filtered.most_common(100))
```

### Observations:

We can now observe that the token list is better and cleaner, with most of the meaningless words and characters removed, giving more meaningful results deduced in the further steps.

```python
# Plotting the frequency distribution of 30 most common words

frequencies_filtered.plot(30,title="Figure 1. Frequency Distribution of 30 Most Common Words")
```

## STEP 5: Sentiment Analysis

```python
# This sentiment analysis section starts by identifying each token's Part of Speech (POS) tag.

from nltk.tag import pos_tag
nltk.download('averaged_perceptron_tagger')

tagged_iphone_tokens = nltk.pos_tag(filtered_tokens)
tagged_iphone_tokens[:20]
```

```python
# Creating a function to extract custom list of tokens based on POS type. For example, list of adjectives, list of nouns, etc.

def create_custom_list(token_list,pos_type):
    custom_list = []
    custom_list = [word 
         for (word, pos) in token_list
         if pos in pos_type] 
    return custom_list
```

```python
# Creating a custom adjectives POS token list

adjective_type = ['JJ','JJR','JJS']
adjectives = create_custom_list(tagged_iphone_tokens,adjective_type)

adjectives[:15]
```

Now we have a list of adjectives used in iPhone 14 reviews. Next, to find out what are the positive and negative comments about the iPhone 14, we will import a dictionary of positive and negative words from external sources and categorize our list of adjectives based on this dictionary.

```python
# Creating a function to read a file and save contents of file in a variable

def read_file(file_name):
    opened_file = open(file_name)
    file_contents = opened_file.read()
    opened_file.close()
    file_contents = file_contents.split(',') # Splitting comma separated string into a list
    return file_contents
```

```python
# Opening and saving the dictionary of positive words file

positive = read_file('./Data/positive-words.txt')
```

```python
# Opening and saving the dictionary of negative words file

negative = read_file('./Data/negative-words.txt')
```

```python
# Analysing adjectives present in our filtered corpus token list to save positive adjectives for positive sentiment analysis

positive_adjectives = [] 

for adj in adjectives:
    for word in positive:
        if adj == word:
            positive_adjectives.append(adj)
            
# Using set to remove duplicate positive tokens
positive_set=set(positive_adjectives)

# Return the format into a list
positive_list=list(positive_set)

# Now preview the list of positive words in our list
positive_list [0:9]
```

```python
# Analysing adjectives present in our filtered corpus token list to save negative adjectives for negative sentiment analysis

negative_adjectives = [] 

for adj in adjectives:
    for word in negative:
        if adj == word:
            negative_adjectives.append(adj)

# Using set to remove duplicate negative tokens
negative_set=set(negative_adjectives)

# Return the format into a list
negative_list=list(negative_set)

# Now preview the list of negative words in our list
negative_list [0:9]
```

To better visualize this list of positive and negative adjectives, we will create two Word Clouds.

```python
# Importing required libraries to create word cloud

!pip install wordcloud
from wordcloud import WordCloud
from PIL import Image
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
    
# Masking image
positive_image = np.array(Image.open('./Image/apple.png'))    
negative_image = np.array(Image.open('./Image/wrench.png'))    

# Creating a function to display word cloud    
def word_cloud(custom_list,cloud_type,shape_mask): 
    cloud = WordCloud(width=800, height=400,background_color="white",max_font_size=160,mask=shape_mask,contour_color='#5A5A5A',contour_width=3,colormap="viridis").generate(' '.join(custom_list))
    
    plt.rcParams["figure.figsize"] = (10,10)
    plt.title(cloud_type + " Word Cloud")
    plt.imshow(cloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()
```

```python
# Word cloud for positive adjectives - PART OF MINI REPORT

cloud_type = "Figure 2. Positive"
word_cloud(positive_list,cloud_type,positive_image)

# Note: The positive sentiment image is visualized in the form of Apple logo
```

```python
# Word Cloud for negative adjectives - PART OF MINI REPORT

cloud_type = "Figure 3. Negative"
word_cloud(negative_list,cloud_type,negative_image)

# Note: The negative sentiment image is visualized in the form of a wrench to illustrate the need for improvement
```

In the last step in this section, we will now measure whether there is more positive or negative sentiment in reviews about the iPhone 14.

```python
positive_adj = len(positive_adjectives)
negative_adj = len(negative_adjectives)

total_adj = positive_adj + negative_adj

positive_response = (positive_adj / total_adj) * 100
negative_response = (negative_adj / total_adj) * 100

print('Number of Positive Adjectives: ' + str(positive_adj))
print ('Positive Response: ' + str(round(positive_response,2)) + ' %')
print('\n')
print('Number of Negative Adjectives: ' + str(negative_adj))
print ('Negative Response: ' + str(round(negative_response,2)) + ' %')
```

**Conclusion**: There is more positive (86%) than negative (14%) sentiment being said about the iPhone 14.

We've looked at the word cloud of positive and negative sentiments for iPhone 14 generated from an adjective list. 

As additional insight, we will now generate a custom list of nouns, which we will use later to see more specifically what is said about iPhone14 in context of the noun, which can be considered as a describe term (or feature) for iPhone 14.

```python
# Creating a custom noun POS token list

noun_type = ['NN','NNS']
nouns = create_custom_list(tagged_iphone_tokens,noun_type)

# Now preview the list of nouns (describe terms) in our list
nouns[:20]
```

## STEP 6: Concordance

```python
# Creating a function to check for concordance

from nltk.text import Text

def concordance_check (corpus_token,word):
    text_of_the_corpus = Text(corpus_token)
    print(text_of_the_corpus.concordance(word))
```

**The purpose of having below code is to allow the client to analyse and check concordance of words of interest as seen from common nouns (considered as describe terms or features) and adjectives (considered as sentiment)**
